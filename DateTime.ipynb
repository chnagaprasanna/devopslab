```python
# Get year, month and date

df=pd.DataFrame({
    "name":["pras","chin"],
    "date":["2/10/2000","08/20/2002"]
})
```

```python
df
```

```python
df.info()
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
df
```

```python
df["date"].dt.year
```

```python
df["year"]=df["date"].dt.year
df["month"]=df["date"].dt.month
df["day"]=df["date"].dt.day

df
```

```python
df["dayoftheyear"]=df["date"].dt.dayofyear
df["dayoftheweek"]=df["date"].dt.day_of_week
df["isleapyear"]=df["date"].dt.is_leap_year
df["quarter"]=df["date"].dt.quarter

df
```

```python
dayweek={
    0:"Monday",
    1:"Tuesday",
    2:"Wednesday",
    3:"Thursday",
    4:"Friday",
    5:"Saturday",
    6:"Sunday"
}
df["day_of_week"]=df["date"].dt.weekday.map(dayweek)
df
```
