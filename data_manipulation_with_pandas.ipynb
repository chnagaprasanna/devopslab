Pivot Tables:
 * alows to reorganize and sumarize selected columns and rows in a dataframe.
 * it first groups and then applies a calculation

```python
import numpy as np
import pandas as pd
data= {
    "A": ["foo","foo","foo","boo","boo","boo"],
    "B": ["one","two","one","two","one","two"],
    "C": ["x","y","x","y","x","y"],
    "D": [1,3,2,4,5,3]
}
df=pd.DataFrame(data)
df
```

```python
pivot_df= df.pivot_table(
    values="D",
    index="A",
    columns="C",
    aggfunc=np.sum
    )

pivot_df
```

```python
pivot_df.reset_index() # reset index
```

Group By:
  * allows to group row data together and call aggregate functions to apply on whole group
  * allows operations includes:
    * Splitting the Object
    * Applying the function
    * Combining the results
  * also in some situations, we split the data into sets and we apply some functionality on each subset.
  * operations includes:
    * Aggregation -> computing a summary statistic.
    * Tranformation -> perform some group specific operation
    * Filteration -> discarding some data with specific condition.
    

```python
pivot_df.groupby(pivot_df.index).agg(np.mean)
```

* Concatenation --> attaches 2 dataframes--> dimensions should be same for both the dataframes.
* Merge --> allows to merge dataframes together using a similar logic as merging SQL tables together.
* Join --> used for combining the columns of different indexed dataframes into a single dataframe.


```python
df1=pd.DataFrame({
    "A": ["A0","A1","A2","A3"],
    "B": ["B0","B1","B2","B3"],
    "C": ["C0","C1","C2","C3"],
    "D": ["D0","D1","D2","D3"]
},index=[0,1,2,3])

df2=pd.DataFrame({
    "A": ["A4","A5","A6","A7"],
    "B": ["B4","B5","B6","B7"],
    "C": ["C4","C5","C6","C7"],
    "D": ["D4","D5","D6","D7"]
},index=[4,5,6,7])

df3=pd.DataFrame({
    "A": ["A8","A9","A10","A11"],
    "B": ["B8","B9","B10","B11"],
    "C": ["C8","C9","C10","C11"],
    "E": ["E8","E9","E10","E11"]
},index=[8,9,10,11])
```

```python
# we can do both inner and outer join
# Parameters("inner", "outer")
# default "outer"
# "outer for union"
# "inner for intersection"

left_df=pd.DataFrame({
    "A": ["A0","A1","A2","A3"],
    "B": ["B0","B1","B2","B3"]
},index=["K0","K1","K2","K3"])

right_df=pd.DataFrame({
    "C": ["C0","C2","C3","C4"],
    "D": ["D0","D2","D3","D4"]
},index=["K0","K2","K3","K4"])

left_df.join(right_df,how="outer")
```

```python
left_df.join(right_df,how="inner")
```

```python
left=pd.DataFrame({
    "K1": ["K0","K0","K1","K2"],
    "K2": ["K0","K1","K0","K1"],
    "A": ["A0","A1","A2","A3"],
    "B": ["B0","B1","B2","B3"]
},index=["K0","K1","K2","K3"])

right=pd.DataFrame({
    "K1": ["K0","K1","K1","K2"],
    "K2": ["K0","K0","K0","K0"],
    "C": ["C0","C2","C3","C4"],
    "D": ["D0","D2","D3","D4"]
},index=["K0","K2","K3","K4"])
```

```python
left
```

```python
right
```

```python
#"inner","left","right"

pd.merge(left,right,how="inner",on=["K1","K2"])
```

```python
pd.merge(left,right,how="left",on=["K1","K2"])
```

```python
pd.merge(left,right,how="right",on=["K1","K2"])
```
